```python
sample_2017 = sample2017[['Respondent','Country','DeveloperType','FormalEducation','MajorUndergrad','Gender','Race']]

sample_2017.fillna('no_answer',inplace=True)

# creating the column 'Trans' to fill with the answers of people who identified themselves as trans or non-binary
sample_2017['Trans'] = sample_2017['Gender'].str.contains('Transgender|Gender non-conforming')

# creating the column 'Year' to identify the respective survey's year
sample_2017['Year'] = 2017

# creating a column to fill with people's sexuality identification
# only valid for 2018, 2019 and 2020 surveys
# 2017 values will be filled with 'no_answer'
sample_2017['SexualityIdentification'] = 'no_answer'

# ordering the columns
sample_2017 = sample_2017[['Respondent','Country','DeveloperType','FormalEducation','MajorUndergrad','Gender','Race','SexualityIdentification','Trans','Year']]

# renaming column names
sample_2017.columns = ['Respondent','Country','DeveloperType','FormalEducation','MajorUndergrad','Gender','Race','Sexuality','Trans','Year']

sample_2017
```

```python
# 2018 survey's columns selection
sample_2018 = sample2018[['Respondent','Country','DevType','FormalEducation','UndergradMajor','Gender','RaceEthnicity','SexualOrientation']]
sample_2018.fillna('no_answer',inplace=True)

# creating the column 'Trans' to fill with the answers of people who identified themselves as trans or non-binary
sample_2018['Trans'] = sample_2018['Gender'].str.contains('Transgender|Non-binary, genderqueer, or gender non-conforming')

# creating the column 'Year' to identify the respective survey's year
sample_2018['Year'] = 2018

# changing the name of the columns
sample_2018.columns = ['Respondent','Country','DeveloperType','FormalEducation','MajorUndergrad','Gender','Race','Sexuality','Trans','Year']

# printing the new data frame
sample_2018
```

```python
# 2019 survey's columns selection
sample_2019 = sample2019[['Respondent','Country','DevType','EdLevel','UndergradMajor','Gender','Ethnicity','Sexuality','Trans']]

# filling the NaN values with a string
sample_2019.fillna('no_answer',inplace=True)

# creating the column 'Year' to identify the respective survey's year
sample_2019['Year'] = 2019

# changing the name of the columns
sample_2019.columns = ['Respondent','Country','DeveloperType','FormalEducation','MajorUndergrad','Gender','Race','Sexuality','Trans','Year']

# printing the new data frame
sample_2019
```

```python
# 2020 survey's columns selection
sample_2020 = sample2020[['Respondent','Country','DevType','EdLevel','UndergradMajor','Gender','Ethnicity','Sexuality','Trans']]

# filling the NaN values with a string
sample_2020.fillna('no_answer',inplace=True)

# creating the column 'Year' to identify the respective survey's year
sample_2020['Year'] = 2020

# changing the name of the columns
sample_2020.columns = ['Respondent','Country','DeveloperType','FormalEducation','MajorUndergrad','Gender','Race','Sexuality','Trans','Year']

# printing the new data frame
sample_2020
```

```python
df_final = pd.concat([sample_2017,sample_2018,sample_2019,sample_2020])
df_final
```

```python
# aggregating female gender
df_final['Gender'] = df_final['Gender'].replace('Woman','Female')
df_final['Gender'] = df_final['Gender'].replace('Female; Transgender','Female')
df_final['Gender'] = df_final['Gender'].replace('Female;Transgender','Female')

# aggregating male gender
df_final['Gender'] = df_final['Gender'].replace('Man','Male')
df_final['Gender'] = df_final['Gender'].replace('Male; Transgender','Male')
df_final['Gender'] = df_final['Gender'].replace('Male;Transgender','Male')
```

```python
df_gender_female = df_final.query('Gender == "Male"| Gender == "Female"').groupby(['Year','Gender'])['Respondent'].count()

df_gender_female = df_gender_female.unstack(level=-1).reset_index()

df_gender_female
```

```python
# creating a column to store the total count of respondents
df_eth_wide['Total'] = df_eth_wide.sum(axis=1)

# creating a column to store the percentual participation of each race
df_eth_wide['%_Asian'] = (df_eth_wide['Asian'] / df_eth_wide['Total']) *100
df_eth_wide['%_Black/African descent'] = (df_eth_wide['Black or of African descent'] / df_eth_wide['Total']) *100
df_eth_wide['%_Hispanic/Latino'] = (df_eth_wide['Hispanic or Latino'] / df_eth_wide['Total']) *100
df_eth_wide['%_Middle Eastern'] = (df_eth_wide['Middle Eastern'] / df_eth_wide['Total']) *100
df_eth_wide['%_Native American/Pacific Islander/Indigenous Australian'] = (df_eth_wide['Native American, Pacific Islander, or Indigenous Australian'] / df_eth_wide['Total']) *100
df_eth_wide['%_White'] = (df_eth_wide['White or of European descent'] / df_eth_wide['Total']) *100
df_eth_wide['%_Multiracial'] = (df_eth_wide['Multiracial'] / df_eth_wide['Total']) *100
df_eth_wide
```
